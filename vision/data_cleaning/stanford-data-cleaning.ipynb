```python
selected_file = df["DDI_file"].tolist()
```

```python
folder_path = "/kaggle/input/ddi-stanford/"
output_folder = "/kaggle/working/imagenet_1pct/benign"

os.makedirs(output_folder, exist_ok=True)

for filename in selected_file:
    
    # Read the image
    image_path = os.path.join(folder_path, filename)
    
    with Image.open(image_path) as img:
        # Convert to JPEG
        jpeg_filename = os.path.splitext(filename)[0] + ".jpeg"
        jpeg_path = os.path.join(output_folder, jpeg_filename)
        img.convert("RGB").save(jpeg_path, "JPEG")
```
